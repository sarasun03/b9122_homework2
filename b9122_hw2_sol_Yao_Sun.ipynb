<h3>Webcrawler 1

```python
from bs4 import BeautifulSoup
import urllib.request

seed_url = "https://www.federalreserve.gov/newsevents/pressreleases.htm"

urls = [seed_url]    
seen = [seed_url]    
opened = []          
covid = [] #create a list for urls that contain the word "covid" on its page

maxNumUrl = 10; #set the minimum number of urls that contains "covid" we want to collect
#print("Starting with url="+str(urls))
while len(urls) > 0 and len(covid) < maxNumUrl:
    
    try:
        curr_url=urls.pop(0)
        #print("num. of URLs in stack: %d " % len(urls))
        #print("Trying to access= "+curr_url)
        req = urllib.request.Request(curr_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()
        opened.append(curr_url)

    except Exception as ex:
        #print("Unable to access= "+curr_url)
        #print(ex)
        continue    

        
    soup = BeautifulSoup(webpage)  
    
    if "covid" in soup.get_text().lower(): #check whether the current url includes "covid"
        covid.append(curr_url)
    
    for tag in soup.find_all('a', href = True): 
        childUrl = tag['href']
        o_childurl = childUrl
        childUrl = urllib.parse.urljoin(seed_url, childUrl)
        #print("seed_url=" + seed_url)
        #print("original childurl=" + o_childurl)
        #print("childurl=" + childUrl)
        #print("seed_url in childUrl=" + str(seed_url in childUrl))
        #print("Have we seen this childUrl=" + str(childUrl in seen))
        if "https://www.federalreserve.gov/newsevents/pressreleases" in childUrl and childUrl not in seen:
            #print("***urls.append and seen.append***")
            urls.append(childUrl)
            seen.append(childUrl)
        else:
            None
            #print("######")

#print("num. of URLs seen = %d, and scanned = %d" % (len(seen), len(opened)))

print("\nThe 10 urls of the webpages found to contain the word “covid”:\n")    
for url in covid:    
    print(url)
```

```python
from bs4 import BeautifulSoup
import urllib.request

seed_url = "https://www.sec.gov/news/pressreleases"

urls = [seed_url]    
seen = [seed_url]    
opened = []          
charges = [] #create a list for urls that contain the word "charges" on its page
text = dict()
text[seed_url] = ""

maxNumUrl = 21; #set the number of urls that contains "charges" we want to collect
#print("Starting with url="+str(urls))
while len(urls) > 0 and len(charges) < maxNumUrl:
    
    try:
        curr_url=urls.pop(0)
        #print("num. of URLs in stack: %d " % len(urls))
        #print("Trying to access= "+curr_url)
        req = urllib.request.Request(curr_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()
        opened.append(curr_url)

    except Exception as ex:
        #print("Unable to access= "+curr_url)
        #print(ex)
        continue    


    soup = BeautifulSoup(webpage)  
    if "charges" in soup.get_text().lower(): #check whether the current url includes "charges"
        charges.append(curr_url)
    
    for tag in soup.find_all('a', href = True): 
        childUrl = tag['href'] 
        txt = tag.get_text() #get the corresponding text of the url
        o_childurl = childUrl
        childUrl = urllib.parse.urljoin(seed_url, childUrl)
        text[childUrl] = txt #append the text to the dictionary
        #print("seed_url=" + seed_url)
        #print("original childurl=" + o_childurl)
        #print("childurl=" + childUrl)
        #print("seed_url in childUrl=" + str(seed_url in childUrl))
        #print("Have we seen this childUrl=" + str(childUrl in seen))
        if "https://www.sec.gov/news/press-release" in childUrl and childUrl not in seen:
            #print("***urls.append and seen.append***")
            urls.append(childUrl)
            seen.append(childUrl)
        else:
            #print("######")
            None

#print("num. of URLs seen = %d, and scanned = %d" % (len(seen), len(opened)))

print("\n20 urls of the webpages found to contain the word “charges”:\n")    
for url in charges[1:]:    
    print(f"url: {url}")
    print(f"text of url: {text[url]}\n")
```
